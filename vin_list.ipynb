```python
import pandas as pd
import hashlib
gen2k_vin_02 = pd.read_csv("../data/vin/gen2k-vin.csv", header = 1, names=["vin", "tcuid", "navi", "tcuid-sub", "navi-sub", "imp-date"],
                       dtype={'vin':'object', 'tcuid':'object', 'navi':'object', 'tcuid-sub':'object','navi-sub':'object',"imp-date":'object'})
gen2k_vin_01 = pd.read_csv("../data/vin/gen2k_01_vin.csv", header = 1, names=["vin", "tcuid", "navi", "tcuid-sub", "navi-sub", "imp-date"],
                       dtype={'vin':'object', 'tcuid':'object', 'navi':'object', 'tcuid-sub':'object','navi-sub':'object',"imp-date":'object'})
#gen2k_vin = gen2k_vin.sort_values('vin')
gen2k_vin_02 = gen2k_vin_02[gen2k_vin_02['tcuid-sub'].isin(['2047'])]
gen2k_vin_01 = gen2k_vin_01[gen2k_vin_01['tcuid-sub'].isin(['2047'])]

gen2k_vin = pd.concat([gen2k_vin_01, gen2k_vin_02])
                           
def mkmd5(str):
    return hashlib.md5(str.encode('utf-8')).hexdigest()
```

```python
from calendar import Calendar
cl = Calendar(firstweekday=6)

gen2k_vin_v = gen2k_vin[['vin']]
gen2k_vin_vd = gen2k_vin_v[~gen2k_vin_v.duplicated()]
gen2k_vin_vd["idx"] = gen2k_vin_vd['vin'].map(mkmd5)
gen2k_vin_vd = gen2k_vin_vd.set_index('idx')

gen2k_vin_v_01 = gen2k_vin_01[['vin']]
gen2k_vin_vd_01 = gen2k_vin_v_01[~gen2k_vin_v_01.duplicated()]
gen2k_vin_vd_01["idx"] = gen2k_vin_vd_01['vin'].map(mkmd5)
gen2k_vin_vd_01 = gen2k_vin_vd_01.rename(columns={'vin':'201901'})
gen2k_vin_vd_01 = gen2k_vin_vd_01.set_index('idx')

gen2k_vin_v_02 = gen2k_vin_02[['vin']]
gen2k_vin_vd_02 = gen2k_vin_v_02[~gen2k_vin_v_02.duplicated()]
gen2k_vin_vd_02["idx"] = gen2k_vin_vd_02['vin'].map(mkmd5)
gen2k_vin_vd_02 = gen2k_vin_vd_02.rename(columns={'vin':'201902'})
gen2k_vin_vd_02 = gen2k_vin_vd_02.set_index('idx')
#gen2k_vin_vi = gen2k_vin_v.rename(columns={'vin':'vini'})
#gen2k_vin_vv = pd.concat([gen2k_vin_v, gen2k_vin_vi], axis=1).set_index('vin')
#gen2k_vin_vd.head()
```

```python
gen2k_vin_vd = pd.concat([gen2k_vin_vd, gen2k_vin_vd_01], axis=1)
for day in cl.itermonthdates(2019, 1):
    try:
        print(day.strftime("%Y%m%d"))
        cl_name = day.strftime("%Y%m%d")
        gen2k_vin_d = gen2k_vin[gen2k_vin['imp-date'].isin([cl_name])]
        gen2k_vin_dv = gen2k_vin_d[['vin']]
        gen2k_vin_dv = gen2k_vin_dv[~gen2k_vin_dv.duplicated()]
        gen2k_vin_dv['idx'] = gen2k_vin_dv['vin'].map(mkmd5)
        gen2k_vin_dv = gen2k_vin_dv.rename(columns={'vin':cl_name}).set_index('idx')
        #gen2k_vin_dv.head()
        #gen2k_vin_vvp = pd.concat([gen2k_vin_vvi,gen2k_vin_vv], axis=1)
        gen2k_vin_vd = pd.concat([gen2k_vin_vd, gen2k_vin_dv], axis=1)
    except:
        print("except:{}".format(day))
gen2k_vin_vd = pd.concat([gen2k_vin_vd, gen2k_vin_vd_02], axis=1)
for day in cl.itermonthdates(2019, 2):
    try:
        print(day.strftime("%Y%m%d"))
        cl_name = day.strftime("%Y%m%d")
        gen2k_vin_d = gen2k_vin[gen2k_vin['imp-date'].isin([cl_name])]
        gen2k_vin_dv = gen2k_vin_d[['vin']]
        gen2k_vin_dv = gen2k_vin_dv[~gen2k_vin_dv.duplicated()]
        gen2k_vin_dv['idx'] = gen2k_vin_dv['vin'].map(mkmd5)
        gen2k_vin_dv = gen2k_vin_dv.rename(columns={'vin':cl_name}).set_index('idx')
        #gen2k_vin_dv.head()
        #gen2k_vin_vvp = pd.concat([gen2k_vin_vvi,gen2k_vin_vv], axis=1)
        gen2k_vin_vd = pd.concat([gen2k_vin_vd, gen2k_vin_dv], axis=1)
    except:
        print("except:{}".format(day))
gen2k_vin_vd.head()
```
